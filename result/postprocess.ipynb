```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
result = np.load("result.npy")
```

```python
params = np.load("params.npy")
```

```python
sign = np.mean(result[:,0,:],axis = -1)
```

```python
Umin, Umax = params[:,1].min(), params[:,1].max()
betamin, betamax = 0.1 * params[:,0].min(), 0.1 * params[:,0].max()
plt.imshow(sign.reshape(5,7),origin = "lower", extent = (Umin, Umax, betamin, betamax), aspect= 'auto')
plt.colorbar()
plt.xlabel("Hubbard U")
plt.ylabel(r"$\beta$")
plt.title(r"$\mu = 1.65$")
```

```python
Number = np.mean(result[:,0,:] * result[:,1,:],axis = -1) / sign
```

```python
doping = (Number - 16 ) /32
```

```python
plt.imshow(doping.reshape(5,7),origin = "lower", extent = (Umin, Umax, betamin, betamax), aspect= 'auto')
plt.colorbar()
plt.xlabel("Hubbard U")
plt.ylabel(r"$\beta$")
plt.title(r"$\mu = 1.65$")
```
